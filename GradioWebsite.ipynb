```python
import gradio as gr
```

```python
import os

def get_chatbot_response(x):
    os.rename(x, x + '.wav')
    return [((x + '.wav',), "What would you like to do with this file?\n 1. Voice Separation\n 2. Singer Information\n 3. Recommendation")]

with gr.Blocks() as demo:
    chatbot = gr.Chatbot()
    mic = gr.Audio(sources=["upload"], type="filepath")
    mic.change(get_chatbot_response, mic, chatbot)

demo.launch()
```
